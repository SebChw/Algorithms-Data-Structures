```python
import itertools
import random 
```

```python
vertices = list(range(1,11))
vertices
```

```python
vertices = list(itertools.combinations(vertices, 2))
random.shuffle(vertices)
print(len(vertices))
vertices = vertices[:int(len(vertices)*0.6)]
print(len(vertices))
```

```python
vertices
```

```python
#now directed graph
vertices2 = list(itertools.permutations(vertices, 2))
random.shuffle(vertices2)
vertices2 = vertices[:int(len(vertices)*0.6)]
vertices2
```
